## Cleaning the Text Data

```python
def remove_html_tags(text):
    pattern="<.*?>"
    clean_text=re.sub(pattern,'',text)
    return clean_text
```

```python
def remove_urls(text):
    pattern='https?://\S+'
    new_text=re.sub(pattern,'',text)
    return new_text
```

```python
df['tweet_content']=df['tweet_content'].apply(remove_urls)
```

```python
df2['tweet_content']=df2['tweet_content'].apply(remove_urls)
```

```python
def remove_newline(text):
    pattern='\n'
    new_text=re.sub(pattern,'',text)
    return new_text
```

```python
df2['tweet_content']=df2['tweet_content'].apply(remove_newline)
```

```python
exclude=list(string.punctuation)
```

```python
def remove_punctuation(text):
    for i in exclude:
        text=text.replace(i,'')
    return text
```

```python
df['tweet_content']=df['tweet_content'].apply(remove_punctuation)
```

```python
import emoji
def handling_emoji(text):
    emoji1 = emoji.emoji_list(text)
    for i in emoji1:
        # print(i['emoji'])
        text=text.replace(i['emoji'],'')
    return text
```

```python
df['tweet_content']=df['tweet_content'].apply(handling_emoji)
```

```python
df2['tweet_content']=df2['tweet_content'].apply(handling_emoji)
```

```python
def text_process(mess):

    STOPWORDS = stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure']
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)

    # Now just remove any stopwords
    return ' '.join([word for word in nopunc.split() if word.lower() not in STOPWORDS])
```

```python

resources = [
    'stopwords',
    'punkt',
    'wordnet',
    'omw-1.4',
    'averaged_perceptron_tagger'
]

for resource in resources:
    nltk.download(resource)
```

```python
df['tweet_content']=df['tweet_content'].apply(text_process)
```

```python
df2['tweet_content']=df2['tweet_content'].apply(text_process)
```

## Tokenization

```python
model_id="bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["tweet_content"],
        truncation=True,
        padding="max_length",
        max_length=128
    )
```

```python
dataset = Dataset.from_pandas(df)
```

```python
dataset2 = Dataset.from_pandas(df2)
```

```python
train_dataset = dataset.map(tokenize_function, batched=True)
test_dataset = dataset2.map(tokenize_function, batched=True)
```

## Raname Column to Labels

```python
train_dataset = train_dataset.rename_column("sentiment", "labels")
test_dataset = test_dataset.rename_column("sentiment", "labels")
```

## Removing the text Column

```python
train_dataset = train_dataset.remove_columns(["tweet_content"])
test_dataset = test_dataset.remove_columns(["tweet_content"])
```

```python
print(type(train_dataset))
```

## Loading a Model

```python
MODEL_NAME = "bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=4      # Positive, Negative, Neutral, Irrelevant
)
```

## Training Configuration

```python
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred

    predictions = np.argmax(logits, axis=-1)

    precision, recall, f1, _ = precision_recall_fscore_support(
        labels,
        predictions,
        average="weighted"
    )

    accuracy = accuracy_score(labels, predictions)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }
```

```python
data_collator = DefaultDataCollator()
```

## Training

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

## Evaluate Results

```python
eval_results = trainer.evaluate()

print(eval_results)
```
